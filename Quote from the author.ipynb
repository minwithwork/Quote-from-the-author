**2.2.4. Thống kê về các câu nói nổi tiếng như: câu dài nhất, ngắn nhất,…**

```python
df['Quote'] = str(df['Quote'])
```

```python
longest_quote = df['Quote'][0]
shortest_quote = df['Quote'][0]
count_word = 0
word_dict = {}

for quote in df['Quote']:
    if( len(longest_quote.split(' ')) < len(quote.split(' ')) ):
        longest_quote = quote
    if( len(shortest_quote.split(' ')) > len(quote.split(' ')) ):
        shortest_quote = quote
    for word in quote.split(' '):
        if( word not in word_dict.keys() ):
            word_dict[word] = 1
        else:
            word_dict[word] += 1
        count_word += 1
        
print('Câu dài nhất: ' + str(len(longest_quote.split(' '))) + ' từ')
print('Câu ngắn nhất: ' + str(len(shortest_quote.split(' '))) + ' từ')
print('Tổng số từ: ' + str(count_word) + ' từ')
print('Tổng số từ vựng: ' + str(len(word_dict.keys())) + ' từ')
```
